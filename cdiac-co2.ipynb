```python
import pandas as pd
import numpy as np

import re
import os
```

```python
def to_concept_id(s):
    '''convert a string to lowercase alphanumeric + underscore id for concepts'''
    s1 = re.sub(r'[/ -\.\*";\n]+', '_', s).lower()

    if s1[-1] == '_':
        s1 = s1[:-1]

    return s1.strip()
```
